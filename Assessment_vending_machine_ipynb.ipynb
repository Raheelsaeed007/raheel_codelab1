```python

#*********************************************VENDING MACHINE***********************************************************
#***********************************************************************************************************************
import time
print("\033[1;31;40m")  # adding red color
# Asking the user how much money they wish to enter.
number_of_dirhams = int(input("How many  dirhams would you like to insert? "))

# Creating a variable to store the total amount of money inserted into the vending machine.
change = round(number_of_dirhams)
# Informing the user how much they have entered in total.
print ("\nIn total you have entered dirhams", change)
time.sleep(2)
# Creating variables for the 6 items and their respective prices. 

item_1 = "Coffee"
price_1 = 4
item_2 = "biscuits"
price_2 = 3.50
item_3 = "Mango juice"
price_3 = 3
item_4 = "cup cake"
price_4 = 2.50
item_5 = "chips"
price_5 = 1.75
item_6 = "chocolate"
price_6 = 2.15
 
items=[item_1,item_2,item_3,item_4,item_5,item_6]
cart = []
# Creating variables to track the number of each items bought,
coffee_bought = 0
biscuits_bought = 0
mango_juice_bought = 0
cup_cake_bought = 0
chips_bought = 0
chocolate_bought = 0
# Informing the user of the choices available and the prices that of each item.
print ("There are 6 items available to pick from;\n")
time.sleep(2)
print (" Item                   Costs")
print (" ----                   -----")
print (" {}         ----->  {} ".format(item_1, price_1))
print (" {}  ----->  {} ".format(item_2, price_2))
print (" {}     ----->  {} ".format(item_3, price_3))
print (" {}       ----->  {} ".format(item_4, price_4))
print (" {}    ----->  {} ".format(item_5, price_5))
print (" {}          ----->  {} ".format(item_6, price_6))
print ("")
# Asking the user to make a selection.
while change > 0:
    customer_choice = input("What would you like to buy? Type N when you are finished \n")
    if customer_choice == "Coffee" or customer_choice == "coffee" and change >= price_1:
        print ("You have chosen a", item_1, "these cost", price_1, "each,")
        change = round((change - price_1),2)
        coffee_bought = (coffee_bought + 1)
        
        print ("You have this much money remaining: dirhams", change)


    elif customer_choice == "biscuits" or customer_choice == "biscuits" and change >= price_2:
        print ("You have chosen a", item_2, "these cost", price_2, "each,")
        change = round((change - price_2),2)
        biscuits_bought = (biscuits_bought + 1)
      
        print ("You have this much money remaining: dirhams", change)


    elif customer_choice == "Mango juice" or customer_choice == "Mango juice" and change >= price_3:
        print ("You have chosen a", item_3, "these cost", price_3, "each,")
        change = round((change - price_3),2)
        Mango_juice_bought = (Mango_juice_bought + 1)
        
        print ("You have this much money remaining: dirhams", change)


    elif customer_choice == "cup cake" or customer_choice == "cup cake" and change >= price_4:
        print ("You have chosen a", item_4, "these cost", price_4, "each,")
        change = round((change - price_4),2)
        cup_cake_bought = (cup_cake_bought + 1)
        
        print ("You have this much money remaining: dirhams", change)


    elif customer_choice == "Chips" or customer_choice == "chips" and change >= price_5:
        print ("You have chosen a", item_5, "these cost", price_5, "each,")
        change = round((change - price_5),2)
        chips_bought = (chips_bought + 1)
        
        print ("You have this much money remaining: dirhams", change)


    elif customer_choice == "chocolate" or customer_choice == "chocolate" and change >= price_6:
        print ("You have chosen a", item_6, "these cost", price_6, "each,")
        change = round((change - price_6),2)
        chocolate_bought = (chocolate_bought + 1)
        
        print ("You have this much money remaining: dirhams", change)


    elif customer_choice == "N" or customer_choice == "n":
       
        break

    elif change <= 0:
        print ("You have run out of money.")
        #using break
        break
    else:
        print ("There has been an error or you may not have enough credit.")
    

        # Informing the user how much they have left to spend.
print ("\nYou have dirhams", change, "remaining.")
time.sleep(2)
# Printing the receipt
print ("\nHere is your receipt;")
time.sleep(2)
if coffee_bought > 0:
    print ("You bought", coffee_bought, item_1, "at dirhams", (price_1 * coffee_bought))
if biscuits_bought > 0:
    print ("You bought", biscuits_bought, item_2, "at dirhams", (price_2 * biscuits_bought))
if mango_juice_bought > 0:
    print ("You bought", mango_juice_bought, item_3, "at dirhams", (price_3 * mango_juice_bought))
if cup_cake_bought > 0:
    print ("You bought", cup_cake_bought, item_4, "at dirhams", (price_4 * cup_cake_bought))
if chips_bought > 0:
    print ("You bought", chips_bought, item_5, "at dirhams", (price_5 * chips_bought))
if chocolate_bought > 0:
    print ("You bought", chocolate_bought, item_6, "at dirhams", (price_6 * chocolate_bought))
print("\033[1;31;40m")  # adding red color
print ("Thanks for using our Vending Machine")







      
      
    

    
```
