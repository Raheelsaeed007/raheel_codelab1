```python
#Chapter 5 Exercise 5 "Pets"


pets=[]

#Charachteristics of pets 

pet={
    'animal' :'Dog',
     'name'  :'max',
     'owner' :'John',
     'weight':14,
     'food'  :'kibble',
}
pets.append(pet)

pet={
    'animal' :'goat',
     'name'  :'rami',
     'owner' :'ramsey',
     'weight':24,
     'food'  :'grass',
}
pets.append(pet)

pet={
    'animal' :'horse',
     'name'  :'philip',
     'owner' :'peter',
     'weight':199,
     'food'  :'forage',
}
pets.append(pet)

for pet in pets:
    print(f"\nAll the information known about {pet['name'].title()}:")
    for key, value in pet.items():
        print(f"\t{key}:{value}")

#All the information about pets and their charachteristics
```
